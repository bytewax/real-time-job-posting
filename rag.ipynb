```python
import json 
from dotenv import load_dotenv
import os
import requests
load_dotenv("source.env")
api_key = os.getenv("api_key")
```

```python

url = "https://jsearch.p.rapidapi.com/search"

querystring = {"query":"AI Engineer","page":"1","num_pages":"15"}

headers = {
    "X-RapidAPI-Key": api_key,
    "X-RapidAPI-Host": "jsearch.p.rapidapi.com"
}


fetched_data = []

response = requests.get(url, headers=headers, params=querystring, timeout=10)
data = response.json()

with open('ai_engineering.json', 'w') as f:
    json.dump(data, f)
```

```python
from jsoncomponent import JsonToDocument
```

```python
converter = JsonToDocument(flatten_field='data', one_doc_per_row=True)
results = converter.run(sources=["ai_engineering.json"])
documents = results["documents"]
```

```python
for i in range(len(documents)):
    print(documents[i].content)
    print("-------------------------")
```
